# House Price Prediction

```python
#setup
import numpy as np
import pandas as pd 
import sklearn
from sklearn.impute import SimpleImputer
```

```python
#Importing training data
train_data = pd.read_csv('C:/Users/44745/Desktop/ML_Resources/projects/IOWA_house_price prediction/train.csv')
```

## Preprocessing

```python
train_data.head()
```

```python
# seperating features and response variables
y = train_data.SalePrice
X = train_data.drop('SalePrice', 1)
X.describe()
```

### Dealing with numeric variables

```python
#checking the numeric variables
X_numeric_cols = X.select_dtypes(include=np.number).columns.tolist()
print(X_numeric_cols)
```

```python
#checking for null values
print('null values before imputing: ')
print(X[X_numeric_cols].isnull().sum())
#imputing the null values using simple Imputer
my_imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputed_df_numeric = pd.DataFrame(my_imputer.fit_transform(X[X_numeric_cols]))
imputed_df_numeric.columns = [X_numeric_cols]
print('null values after imputing: ')
print(imputed_df_numeric.isnull().sum())
print('summary: ')
imputed_df_numeric = imputed_df_numeric.drop('Id',1)
imputed_df_numeric.head(6)
```

### Dealing with Categorical Variable

```python
X_cat_cols = X.select_dtypes(include='object').columns.tolist()
print(X_cat_cols)
```

```python
#checking for missing data
print(X[X_cat_cols].isnull().sum())
#percentage of missing values in each column
percent_missing = X[X_cat_cols].isnull().sum() * 100 / len(X)
missing_value_df = pd.DataFrame({'column_name': X[X_cat_cols].columns,
                                 'percent_missing': percent_missing}).reset_index(drop = True)
print("percentage of missing data: ")
print(missing_value_df.head(len(missing_value_df)))
```

```python
#selecting variables to drop due to missing values
index_to_drop = []
for i in range(len(missing_value_df)):
    if missing_value_df.at[i, 'percent_missing'] > 50:
        ind = missing_value_df.at[i, 'column_name']
        index_to_drop.append(ind)
print(index_to_drop)
```

```python
#Dropping the columns not required for assessment
X_cat = X.loc[:,X_cat_cols]
X_cat = X.drop(index_to_drop,axis = 1)
```

```python
#imputing the missing value and encoding the variables
from sklearn.preprocessing import OneHotEncoder
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imputed_df_cat = pd.DataFrame(imputer.fit_transform(X_cat))
imputed_df_cat.columns = X_cat.columns
OH_encoder = OneHotEncoder(handle_unknown = 'ignore', sparse = False)
encoded_cat_df = pd.DataFrame(OH_encoder.fit_transform(imputed_df_cat))
encoded_cat_df.head()
```

```python
#concatenating the entire dataframe
X_train_final = pd.concat([imputed_df_numeric, encoded_cat_df], axis = 1)
```

```python
#splitting training data into training and validation set
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X_train_final, y, random_state = 1)
```

```python
#creating model
from xgboost import XGBRegressor
model = XGBRegressor(n_estimators = 500, learning_rate = 0.05)

model.fit(train_X, train_y, early_stopping_rounds=5, eval_set = [(test_X, test_y)], verbose = False)
```
